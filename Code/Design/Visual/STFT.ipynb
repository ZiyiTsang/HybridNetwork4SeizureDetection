```python
import pandas as pd
import re
import os
import numpy as np
project_path = os.path.abspath(os.path.relpath('../../../../', os.getcwd()))

STFT_path=os.path.join(project_path,'BilinearNetwork/Data/PreprocessedData/CHB-MIT/detection2s')
```

```python
# 读取STFT数据
stft_data=np.load(os.path.join(STFT_path,'chb01-cross-1.npz'))['X']
stft_data.shape
```

```python
import numpy as np


def batch_zscore_normalize(batch_array):
    normalized_batch_array = np.zeros_like(batch_array)
    for i in range(batch_array.shape[0]):
        sample = batch_array[i]
        mean_val = np.mean(sample)
        std_dev = np.std(sample)
        normalized_sample = (sample - mean_val) / std_dev
        normalized_batch_array[i] = normalized_sample
    return normalized_batch_array
stft_data_=batch_zscore_normalize(stft_data)
```
