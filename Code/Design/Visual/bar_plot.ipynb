```python
plt.figure(figsize=(8, 4.5),dpi=300)
data = {
    'Model': ['Conv-Small', 'Conv-Large', 'Rec-Small', 'Rec-Large','Transf-Small', 'Transf-Large','Conv-Small & Conv-Large', 'Conv-Large & Transf-Large','Conv-Large & Transf-Small','Rec-Small &Transf-Small','Rec-Large & Transf-Small'],
    'Mean': [4.63, 3.96, 3.91, 3.83, 3.8, 3.78, 3.82, 3.72, 4.09,3.85,4.14],
    'Std': [3.61, 2.37, 2.5, 2.25, 2.61, 2.49, 2.59, 2.11, 2.27,2.55,2.73],
    'Group': ['Individual FE', 'Individual FE', 'Individual FE', 'Individual FE', 'Individual FE', 'Individual FE', 'Hybrid FE', 'Hybrid FE', 'Hybrid FE', 'Hybrid FE', 'Hybrid FE']
}
data= pd.DataFrame(data)
sns.barplot( x='Model', y='Mean',  data=data, capsize=10,hue='Group',palette='Set2',dodge=True,width=1.6)
plt.xticks(rotation=30)
plt.tight_layout()


plt.xlim(-1, 12)
plt.ylabel('Detection Latency (s)')
plt.xlabel('')
plt.legend(title='Group', loc='upper right')
plt.ylim(3.6, 4.7)


current_ticks = plt.gca().get_xticks()
current_labels = plt.gca().get_xticklabels()


new_ticks = current_ticks - 0.5
plt.gca().set_xticks(new_ticks)
plt.gca().set_xticklabels(data['Model'])
sns.despine()


plt.axhline(y=3.78, xmin=0.405,xmax=1,color='purple', linestyle='--', linewidth=1)
plt.axhline(y=3.72, xmin=0.62,xmax=1,color='purple', linestyle='--', linewidth=1)

plt.show()
```
